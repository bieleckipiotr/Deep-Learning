```python
import numpy as np
import tensorflow as tf
from torchvision import transforms
from PIL import Image as im
import glob
```

```python
target_dir = 'data/data0/lsun/bedroom/2' #choose the folder
files = []
for f in glob.glob(target_dir + '/**/*', recursive=True):
    if f[-4:] == '.jpg':
        print(f)
        files.append(f)
```

```python
transform = transforms.Compose([
    transforms.Resize((64, 64))
])

train_images = []

for i, file in enumerate(files, 1):
    image_loaded = tf.keras.utils.load_img(file)
    train_images.append(np.array(transform(image_loaded)))
    if i % 30000 == 0 or i == len(files):
        train_images = np.array(train_images).astype('float32')
        train_images = (train_images - 127.5) / 127.5
        file_name = f"train_images_{2}.npy" #change the number here if needed
        np.save(file_name, train_images)
        print(f"Saved to {file_name}")
        train_images = []
```
